# Timedelta & Date Arithmetic

```python
start = date(2025, 1, 10)
end = date(2025, 2, 5)
duration = (end - start).days
print(f'The duration is {duration} days')
```

```python
days_after_start = start + timedelta(days=10)
days_before_start = start - timedelta(days=14)
weeks_after_start = start + timedelta(weeks=3)
the_day = datetime(2025, 2, 18, 11, 23, 10) + timedelta(hours=5, minutes=20, seconds=45)
the_day.strftime("%B %d, %Y at %I:%M:%S %p")
```

# Date in Pandas

```python
df_dates = pd.DataFrame({
    "customer_id": [101, 102, 103, 104, 105],
    "signup_date": ["2024-01-15", "2024-02-20", "2024-03-05", "2024-03-18", "2024-04-10"],
    "last_purchase": ["2024-06-01", "2024-06-15", "2024-07-01", "2024-07-08", "2024-07-12"]
})

df_dates
```

```python
df_dates.dtypes
```

```python
df_dates["signup_date"] = pd.to_datetime(df_dates["signup_date"])
df_dates["last_purchase"] = pd.to_datetime(df_dates["last_purchase"])
df_dates.dtypes
```

```python
df_dates["signup_year"] = df_dates["signup_date"].dt.year
df_dates["signup_month"] = df_dates["signup_date"].dt.month
df_dates["signup_day"] = df_dates["signup_date"].dt.day
df_dates["last_purchase_year"] = df_dates["last_purchase"].dt.year
df_dates["last_purchase_month"] = df_dates["last_purchase"].dt.month
df_dates["last_purchase_day"] = df_dates["last_purchase"].dt.day
df_dates
```

## Day of the week & weekend flag

```python
df_dates['signup_dow'] = df_dates['signup_date'].dt.dayofweek
df_dates['is_weekend'] = df_dates['signup_dow'].isin([5, 6])    
df_dates
```

## `days_between_signup_and_purchase`

```python
df_dates['days_between_signup_and_purchase'] = (df_dates['last_purchase'] - df_dates['signup_date']).dt.days
df_dates
```

## Filter Data

```python
df_dates[df_dates['signup_date'] > "2024-03-01"]
```

```python
df_dates[df_dates['signup_date'].dt.month == 3]
df_dates[df_dates['signup_date'].dt.year == 2024]
```

## Sort by signup date

```python
df_dates.sort_values(by='signup_date')
```

## Date ranges

```python
start_date = pd.to_datetime("2024-07-01")
end_date = start_date + timedelta(days=120)
pd.date_range(start=start_date, end=end_date, freq='10D')
pd.date_range(start=start_date, freq='D', periods=10)
```

```python
start_date = pd.to_datetime("2024-01-01")
pd.date_range(start=start_date, freq='MS', periods=12)
```

## Data Frame with date range
### Derived columns:
1. year
2. month
3. day
4. day_name
5. dow
6. is_weekend

```python
df_new_dates=pd.DataFrame({
    'date': pd.date_range(start=start_date, freq='D', periods=15),
    'sales': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800]
    })
df_new_dates
```

```python
df_new_dates['year'] = df_new_dates['date'].dt.year
df_new_dates['month'] = df_new_dates['date'].dt.month
df_new_dates['day'] = df_new_dates['date'].dt.day
df_new_dates['dow'] = df_new_dates['date'].dt.day_of_week
df_new_dates['day_name'] = df_new_dates['date'].dt.day_name()
df_new_dates['is_weekend'] = df_new_dates['day_name'].isin(['Saturday', 'Sunday'])
df_new_dates
```

## Beginning and End of Month for One Date

```python
some_date = pd.Timestamp("2024-08-18")
print(f'{some_date.year} is the year')
print(f'{some_date.month} is the month')
print(f'{some_date.day} is the day')
print(f'{some_date.hour} is the hour')
print(f'{some_date.minute} is the minute')
print(f'{some_date.second} is the second')
print(f'{some_date.date()} is the date')
```

```python
print(f'{some_date.to_period("M").start_time.date()} is the month\'s start')
print(f'{some_date.to_period("M").end_time.date()} is the month\'s end')
```
